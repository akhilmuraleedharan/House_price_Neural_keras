### We  decide on the features wee can adopt for the model we propose

```python
data_feature =  data[["price","bedrooms","bathrooms","sqft_living","sqft_lot","sqft_above","sqft_basement","yr_built"]]
```

```python
sns.pairplot(data_feature)
```

```python
selected_features = ['bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 'sqft_above', 'sqft_basement']
```

We divide the data into the independent and target variables

```python
X =  data[selected_features]
```

```python
X.head()
```

```python
Y =  data["price"]
```

```python
Y
```

### Use a minmax scaler to transform the values 

```python
from sklearn.preprocessing import MinMaxScaler
scaler =  MinMaxScaler()

X_scaled =  scaler.fit_transform(X)
```

```python
X_scaled
```

```python
Y = Y.values.reshape(-1,1)
y_scaled =  scaler.fit_transform(Y)
```

```python
y_scaled
```

### Split the data into test and train portions

```python
from sklearn.model_selection import  train_test_split

X_train,X_test,y_train, y_test =  train_test_split(X_scaled,y_scaled,test_size=0.25, random_state=42)
```

### We import the sequential and dense modules from the keras library

```python
import tensorflow.keras

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense 
```

```python
n_cols = X_train.shape[1]
```

```python
n_cols
```

### We define a neural network function which is used to model the data

```python
def regression_model():
    model = Sequential()
    model.add(Dense(300, activation='relu',input_shape=(n_cols,)))
    model.add(Dense(300,activation='relu'))
    model.add(Dense(300,activation='relu'))
    model.add(Dense(1))

    model.compile(optimizer= 'adam', loss = 'mean_squared_error')
    return model
```

```python
model =  regression_model()
```

```python
model 
```

```python
model.summary()
```

```python
epochs_hist = model.fit(X_train, y_train, epochs = 100, batch_size = 50, validation_split = 0.2)
```

```python
epochs_hist.history.keys()
```

Plot the training and validation loss

```python
plt.plot(epochs_hist.history['loss'])
plt.plot(epochs_hist.history['val_loss'])
plt.title('Model Loss Progress During Training')
plt.xlabel('Epoch')
plt.ylabel('Training and Validation Loss')
plt.legend(['Training Loss', 'Validation Loss'])
```

### We the use the model on the test data

```python
X_test_1 = np.array([[ 4, 3, 1960, 5000, 1, 2000, 3000 ]])

scaler_1 = MinMaxScaler()
X_test_scaled_1 = scaler_1.fit_transform(X_test_1)

y_predict_1 = model.predict(X_test_scaled_1)

y_predict_1 = scaler.inverse_transform(y_predict_1)
y_predict_1
```

```python
y_predict = model.predict(X_test)
plt.plot(y_test, y_predict, "^", color = 'r')
plt.xlabel('Model Predictions')
plt.ylabel('True Values')
```

Ideally all tha values should be on a 45 degree line if the predictions are accurate. here we see that the plot is a cone with variations from the 45 degree angle. We will also check the accuracy of the model

```python
n = len(X_test)
k = X_test.shape[1]
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
from math import sqrt

RMSE = float(format(np.sqrt(mean_squared_error(y_test, y_predict)),'.3f'))
MSE = mean_squared_error(y_test, y_predict)
MAE = mean_absolute_error(y_test, y_predict)
r2 = r2_score(y_test, y_predict)
adj_r2 = 1-(1-r2)*(n-1)/(n-k-1)

print('RMSE =',RMSE, '\nMSE =',MSE, '\nMAE =',MAE, '\nR2 =', r2, '\nAdjusted R2 =', adj_r2) 
```

### Here we see that the  model is not that accurate according to the scoring metrics. We will now adopt more features to check if the scoring metrics of the model can be improved

```python
selected_features = ['bedrooms','bathrooms','sqft_living','sqft_lot','floors', 'sqft_above', 'sqft_basement', 'waterfront', 'view', 'condition', 'grade', 'sqft_above', 'yr_built', 
'yr_renovated', 'zipcode', 'lat', 'long', 'sqft_living15', 'sqft_lot15']

X = data[selected_features]
```

```python
X_scaled = scaler.fit_transform(X)
```

```python
y = data['price']
```

```python
y = y.values.reshape(-1,1)
y_scaled = scaler.fit_transform(y)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_scaled, test_size = 0.25, random_state=42)
```

```python
n_cols = X_train.shape[1]
```

### We have added one more neural network to improve the predictions from  the model

```python
def regression_model():
    model = Sequential()
    model.add(Dense(300, activation='relu',input_shape=(n_cols,)))
    model.add(Dense(300,activation='relu'))
    model.add(Dense(300,activation='relu'))
    model.add(Dense(300,activation='relu'))
    model.add(Dense(1))

    model.compile(optimizer= 'adam', loss = 'mean_squared_error')
    return model
```

```python
model  = regression_model()
```

```python
epochs_hist = model.fit(X_train, y_train, epochs = 100, batch_size = 50, validation_split = 0.2)
```

```python
plt.plot(epochs_hist.history['loss'])
plt.plot(epochs_hist.history['val_loss'])
plt.title('Model Loss Progress During Training')
plt.ylabel('Training and Validation Loss')
plt.xlabel('Epoch number')
plt.legend(['Training Loss', 'Validation Loss'])
```

```python
y_predict = model.predict(X_test)
plt.plot(y_test, y_predict, "^", color = 'r')
plt.xlabel("Model Predictions")
plt.ylabel("True Value (ground Truth)")
plt.title('Linear Regression Predictions')
plt.show()
```

### As can be seen from the figure, the model accuracy has improved when compared to the previous case

```python
y_predict_orig = scaler.inverse_transform(y_predict)
y_test_orig = scaler.inverse_transform(y_test)
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
from math import sqrt

RMSE = float(format(np.sqrt(mean_squared_error(y_test_orig, y_predict_orig)),'.3f'))
MSE = mean_squared_error(y_test_orig, y_predict_orig)
MAE = mean_absolute_error(y_test_orig, y_predict_orig)
r2 = r2_score(y_test_orig, y_predict_orig)
adj_r2 = 1-(1-r2)*(n-1)/(n-k-1)

print('RMSE =',RMSE, '\nMSE =',MSE, '\nMAE =',MAE, '\nR2 =', r2, '\nAdjusted R2 =', adj_r2) 
```

## The scoring metrics also indicate that the model has better prediction capabilities
